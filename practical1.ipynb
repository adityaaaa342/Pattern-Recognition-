```python
import pandas as pd
import numpy as np
from sklearn.metrics.pairwise import cosine_similarity
```

```python
customers = pd.DataFrame({
    "Customer": ["C1", "C2", "C3", "C4", "C5"],
    "Annual_Spend": [5000, 5500, 7000, 4800, 6500],
    "Purchase_Frequency": [20, 22, 30, 18, 28],
    "Rating": [4.5, 4.7, 4.8, 4.2, 4.6],
    "Browsing_Time": [60, 65, 80, 55, 75]
})

customers
```

```python
features = customers.iloc[:,1:].values

print(features)
```

```python
for i in range(len(customers)):
    for j in range(i+1,len(customers)):
        dist = np.linalg.norm(features[i]-features[j])
        print(customers["Customer"][i],"-",customers["Customer"][j],"=",round(dist,2))
```

```python
cos_sim = cosine_similarity(features)

for i in range(len(customers)):
    for j in range(i+1,len(customers)):
        print(customers["Customer"][i],"-",customers["Customer"][j],"=",round(cos_sim[i][j],4))
```

```python
sequences = {
    "C1":"ELECTRONICSBOOKS",
    "C2":"ELECTRONICSGAMES",
    "C3":"BOOKSCLOTHING",
    "C4":"ELECTRONICSBOOKS",
    "C5":"BOOKSGAMES"
}

for c,s in sequences.items():
    print(c,":",s)
```

```python
def edit_distance(s1,s2):

    m,n=len(s1),len(s2)

    dp=[[0]*(n+1) for _ in range(m+1)]

    for i in range(m+1):
        dp[i][0]=i

    for j in range(n+1):
        dp[0][j]=j

    for i in range(1,m+1):
        for j in range(1,n+1):

            if s1[i-1]==s2[j-1]:
                dp[i][j]=dp[i-1][j-1]
            else:
                dp[i][j]=1+min(
                    dp[i-1][j],
                    dp[i][j-1],
                    dp[i-1][j-1]
                )

    return dp[m][n]
```

```python
keys=list(sequences.keys())

for i in range(len(keys)):
    for j in range(i+1,len(keys)):
        d=edit_distance(sequences[keys[i]],sequences[keys[j]])
        print(keys[i],"-",keys[j],"=",d)
```
